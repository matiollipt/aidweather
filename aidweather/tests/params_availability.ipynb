```python
from core import AidWeather
from config import cfg


params = list(cfg["weather_params_default"].keys())
print(len(params))
print(params)
```

## Test params availability

```python
for p in params:
    try:
        weather = AidWeather("test", 0, 0, "2020-01-01", "2020-01-05", params=[p])
        df = weather.get()
        assert len(df) == 5
        print(f"{p} params OK")

    except Exception as e:
        print(f"{p} params failed → {e}")
```

## Test how many parameters can be fetched

```python
for i in range(5, len(params) + 1):
    try:
        subset = params[:i]
        weather = AidWeather("test", 0, 0, "2020-01-01", "2020-01-05", params=subset)
        df = weather.get()
        print(f"{i} params OK")
    except Exception as e:
        print(f"{i} params failed → {e}")
        break
```

```python
def test_fetch():
    weather = AidWeather("test", 0, 0, "2020-01-01", "2020-01-31", params=params)
    df = weather.get()
    assert len(df) == 31
    display(df.shape, df.head(), df.tail())

test_fetch()
```

```python
def test_filter():
    # fetch data
    weather = AidWeather("test", 0, 0, "2020-01-01", "2020-01-31", params=params)
    df = weather.get()
    assert len(df) == 31
    
    # filter data
    df = weather.filter(start="2020-01-01", end="2020-01-15")
    assert len(df) == 15
    display(df.shape, df.head(), df.tail())
    
test_filter()
```
